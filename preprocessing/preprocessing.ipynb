```python
def IQR(data, feature_names):

    iqr = pd.DataFrame(0, index=feature_names, columns=['lower', 'min', 'num_lower', 'upper', 'max', 'num_upper', 'percentage'])

    for name in feature_names:
        (Q1, Q3) = data[name].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        upper = Q3 + (1.5 * IQR)
        lower = Q1 - (1.5 * IQR)

        iqr.loc[name, 'upper'] = upper
        iqr.loc[name, 'lower'] = lower

        num_of_out_lower = (data[name] < lower).sum()
        num_of_out_upper = (data[name] > upper).sum()

        percentage = "{:.2f}%".format((num_of_out_lower + num_of_out_upper) / data.shape[0] * 100)

        iqr.loc[name, 'num_lower'], iqr.loc[name, 'num_upper'], iqr.loc[name, 'percentage']  = num_of_out_lower, num_of_out_upper, percentage
        iqr.loc[name, 'min'], iqr.loc[name, 'max'] = min(data[name]), max(data[name])
    return iqr
```

```python
iqr_outliers = IQR(X_train, X_train.columns)
```

```python
def remove_outliers(df, feature_names):
    for name in feature_names:
        (Q1, Q3) = df[name].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        upper = Q3 + (1.5 * IQR)
        lower = Q1 - (1.5 * IQR)
        df = df[(df[name] > lower) & (df[name] < upper)]
    return df

def replace_outliers(df, feature_names):
    for name in feature_names:
        (Q1, Q3) = df[name].quantile([0.25, 0.75])
        IQR = Q3 - Q1
        upper = Q3 + (1.5 * IQR)
        lower = Q1 - (1.5 * IQR)
        df[name] = df[name].apply(lambda x: upper if x > upper else lower if x < lower else x)
    return df
```

```python
#Processing outliers only from train set
```

```python
X_train_no_outliers = remove_outliers(X_train.copy(), X_train.columns)
y_train_no_outliers = y_train[X_train_no_outliers.index]
```

```python
X_train_replaced = replace_outliers(X_train.copy(), X_train.columns)
y_train_replaced = y_train.copy()
```

```python
dump((X_train_no_outliers, y_train_no_outliers), 'no_outliers_train.joblib')
dump((X_train_replaced, y_train_replaced), 'replaced_outliers_train.joblib')
```
